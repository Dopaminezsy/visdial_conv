```python

visdial_conv_dense_path = "../../data/crowdsourced/visdial_1.0_val_dense_annotations_crowdsourced.json"
vispro_dense_path = "../../data/vispro/visdial_1.0_val_dense_annotations_vispro.json"
```

```python
def json_load(file_path):
    with open(file_path, "r") as fb:
        data = json.load(fb)
    return data
```

```python
dense_visdial_conv = json_load(visdial_conv_dense_path)
dense_vispro = json_load(vispro_dense_path)
```

```python
print(len(dense_visdial_conv))
print(len(dense_vispro))
```

```python
vispro_ids = [dense_eg['image_id'] for dense_eg in dense_vispro]

num_vispro_examples = len(vispro_ids)
print("Total vispro examples: ", len(vispro_ids))

visdialconv_ids = [dense_eg['image_id'] for dense_eg in dense_visdial_conv]

print("Total visdialconv examples: ", len(visdialconv_ids))

num_visdialconv_examples = len(visdialconv_ids)
```

```python
def intersection(lst1, lst2): 
    return list(set(lst1) & set(lst2)) 


common_ids = intersection(vispro_ids, visdialconv_ids)

print("Total common examples: ", len(common_ids))

num_common_examples = len(common_ids)


print("Total vispro examples: ", len(vispro_ids))
print("Total visdialconv examples: ", len(visdialconv_ids))

print("Common % wrt vispro: ", round(num_common_examples/num_vispro_examples*100, 2))
print("Common % wrt visdialconv: ", round(num_common_examples/num_visdialconv_examples*100, 2))
```
